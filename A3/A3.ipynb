```python
import folium
```

```python
data = [("Schnaitl Bar", 47.8030008,13.0435397), ("Techno Z",47.8233907,13.0400552)]
data_coordinates = [(47.8030008,13.0435397), (47.8233907,13.0400552)]
```

```python
m = folium.Map(location=[47.8053133,13.0410201],zoom_start=13)
m
```

```python
folium.Marker(location = [data[0][1], data[0][2]],
             popup = data[0][0]).add_to(m)
folium.Marker(location = [data[1][1], data[1][2]],
             popup = data[1][0]).add_to(m)
m
```

```python
my_PolyLine=folium.PolyLine(locations= data_coordinates, weight = 2)
m.add_child(my_PolyLine)
```
